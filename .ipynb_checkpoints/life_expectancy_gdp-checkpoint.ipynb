Section for importing necessary modules

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from scipy.stats import ttest_1samp
from sklearn.linear_model import LinearRegression
#import statsmodels.api as sm
import seaborn as sns
import numpy as np
```

Create dataframes

```python
data = pd.read_csv('all_data.csv')
global_census = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_2252106_cleaned.csv')
```

starting workspace and summaries - basic fiddling with new dataframes

```python
print(data.head())
print(data.describe())
print(data['Country'].unique())
print(global_census.head())
```

The Global Census data is good, but there's too much of it.  Let's pare it down to match with our working dataset for this project and then attach it on as a new column.

```python
countries = data['Country'].unique()
countries = countries.tolist()
countries.append('United States')
#Leaving these here.  There must be some way to use this list in value_vars.  Need to come back and tinker with this later.
years = data['Year'].unique()
years = years.tolist()

regional_census = global_census[global_census['Country Name'].isin(countries)]
#print(regional_census.head(6))
census = pd.melt(frame=regional_census, id_vars=['Country Name'], value_vars=['2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015'], value_name='Population', var_name='Year')
#print(census.head(10))
census['Year'] = census['Year'].astype(int)
census=census.rename(columns={'Country Name': 'Country'})
#print(data.dtypes)
#print(census.dtypes)
census.loc[census['Country']=='United States', 'Country']='United States of America'
data=pd.merge(data, census)
data.loc[data['Country']=='United States of America', 'Country']='United States'
#old name used when I was doublechecking that my merge worked.
#print(data_trial.head())
#print(census.head())
#print(data_trial.describe())
#print(census['Country'].unique())
#print(data_trial['Country'].unique())
```

Next we'll add on some more variables I want to look at.  Namely, GDP Per Capita, as well as percentage changes in GDP, Life Expectancy, Population, and GDP/Capita.

```python
#Let's start by getting GDP per capita.
data['GDP Per Capita'] = data['GDP']/data['Population']

#Setting up a function to create columns by annual change.
def annual_change(row, column):
    if row['Year']==2000:
        return 0
    return ((row[column]/data[(data['Country']==row['Country']) & (data['Year']==row['Year']-1)][column].sum())-1)*100

#Now adding them on
data['Life Expectancy Per Annum'] = data.apply(lambda row: annual_change(row, 'Life expectancy at birth (years)'), axis=1)
data['GDP Per Annum'] = data.apply(lambda row: annual_change(row, 'GDP'), axis=1)
data['GDP Per Capita Per Annum'] = data.apply(lambda row: annual_change(row, 'GDP Per Capita'), axis=1)

#print(data.head(10))

# Now that all of the data is in place, let's get some preliminary looks at it.
print(data.describe())
for country in data['Country'].unique():
    print('\n\n')
    print(country)
    print(data[data['Country']==country].describe())
#print(data.groupby(['Country']).agg({'GDP Per Capita': ['mean'], 'GDP': ['mean']}))
```

Now that the data is wrangled.  Let's move on to properly exploring it.  This will involve making some line graphs of the per Annum changes showing Life Expectancy paired against GDP.  As well as violin plots of the raw numbers to show the shape of them.  Then we'll engage some hypothesis testing on the various variables in order to see if any meaningful numbers appear.  These will all be done both at a 'global' level as well as at the national level to see if any of them show interesting results.

At this point, having seen the summaries, my instinct says that GDP will not over all appear to be connected to life expectancy.

```python
corr, p = pearsonr(data['Life expectancy at birth (years)'], data['GDP'])
print('Pearsons correlation: %.3f' % corr)
corr_use, p = pearsonr(data['Life expectancy at birth (years)'], data['GDP Per Capita'])
print('Pearsons correlation: %.3f' % corr_use)
corr, p = pearsonr(data['Life Expectancy Per Annum'], data['GDP Per Annum'])
print('Pearsons correlation: %.3f' % corr)
corr, p = pearsonr(data['Life Expectancy Per Annum'], data['GDP Per Capita Per Annum'])
print('Pearsons correlation: %.3f' % corr)

countries = data['Country'].unique()
gdp_corr = []
gdppc_corr = []
for country in countries:
    print('{}:'.format(country))
    corr, p = pearsonr(data[data['Country']==country]['Life expectancy at birth (years)'], data[data['Country']==country]['GDP'])
    print('Pearsons correlation: %.3f' % corr)
    gdp_corr.append(corr)
    corr, p = pearsonr(data[data['Country']==country]['Life expectancy at birth (years)'], data[data['Country']==country]['GDP Per Capita'])
    print('Pearsons correlation: %.3f' % corr)
    gdppc_corr.append(corr)
    print('Median Life Expectancy: {}'.format(data[data['Country']==country]['Life expectancy at birth (years)'].median()))
    print('Median GDP Per Capita: {}'.format(data[data['Country']==country]['GDP Per Capita'].median()))
    print('\n')

print(data.groupby('Country').agg({'Life expectancy at birth (years)': ['median', 'mean','max', 'min'], 'GDP': ['median', 'mean'], 'GDP Per Capita': ['median', 'mean', 'min', 'max']}))

print('\nAverage correlation of national correspondence: GDP - {}'.format(np.mean(gdp_corr)))
print('Average correlation of national correspondence: GDP Per Capita - {}'.format(np.mean(gdppc_corr)))
tstat, pval = ttest_1samp(gdp_corr, corr_use)
print('P Value of national vs. global correspondence: {}'.format(pval))
```

```python
corr, p = pearsonr(data['Life expectancy at birth (years)'], data['GDP'])
print('Pearsons correlation: %.3f' % corr)
corr, p = pearsonr(data['Life expectancy at birth (years)'], data['GDP Per Capita'])
print('Pearsons correlation: %.3f' % corr)
corr, p = pearsonr(data['Life Expectancy Per Annum'], data['GDP Per Annum'])
print('Pearsons correlation: %.3f' % corr)
corr, p = pearsonr(data['Life Expectancy Per Annum'], data['GDP Per Capita Per Annum'])
print('Pearsons correlation: %.3f' % corr)

countries = data['Country'].unique()
colours = ['grey', 'red', 'black', 'green', 'blue', 'goldenrod']
fig = plt.figure(figsize=(20, 20))

ax1 = fig.add_subplot(2, 2, 1)
for country, colour in zip(countries, colours):
    ax1.scatter(data[data['Country']==country]['Life expectancy at birth (years)'], data[data['Country']==country]['GDP'], alpha=0.8, c=colour, edgecolors='none', s=30, label=country)

plt.title('GDP and Life Expectancy by Country')
plt.legend(loc=2)
plt.xlabel('Life Expectancy (Years)')
plt.ylabel('GDP ($10T)')

ax2 = fig.add_subplot(2, 2, 2)
for country, colour in zip(countries, colours):
    ax2.scatter(data[data['Country']==country]['Life expectancy at birth (years)'], data[data['Country']==country]['GDP Per Capita'], alpha=0.8, c=colour, edgecolors='none', s=30, label=country)

plt.title('GDP Per Capita and Life Expectancy by Country')
plt.legend(loc=2)
plt.xlabel('Life Expectancy (Years)')
plt.ylabel('GDP Per Capita ($)')

ax3 = fig.add_subplot(2, 2, 3)
for country, colour in zip(countries, colours):
    ax3.scatter(data[data['Country']==country]['Life Expectancy Per Annum'], data[data['Country']==country]['GDP Per Annum'], alpha=0.8, c=colour, edgecolors='none', s=30, label=country)

plt.title('GDP and Life Expectancy Changes Per Year by Country')
plt.legend(loc=2)
plt.xlabel('Change in Life Expectancy from Previous Year (%)')
plt.ylabel('Change in GDP from Previous Year (%)')

ax4 = fig.add_subplot(2, 2, 4)
for country, colour in zip(countries, colours):
    ax4.scatter(data[data['Country']==country]['Life Expectancy Per Annum'], data[data['Country']==country]['GDP Per Capita Per Annum'], alpha=0.8, c=colour, edgecolors='none', s=30, label=country)

plt.title('GDP Per Capita and Life Expectancy Changes Per Year by Country')
plt.legend(loc=2)
plt.xlabel('Change in Life Expectancy from Previous Year (%)')
plt.ylabel('Change in GDP Per Capita from Previous Year (%)')
plt.savefig('Life Expectancy Scatterplots.png')
plt.show()
```
